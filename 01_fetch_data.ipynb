Imports

```python
import pandas as pd
import json
import requests as req
```

Constants

```python
API_URL = "https://recruitment.aimtechnologies.co/ai-tasks"
```

Read dialect dataset file

```python
dialect_df = pd.read_csv('dialect_dataset.csv')
dialect_df
```

Explore dialect file

```python
"Available Classes : ",dialect_df.dialect.unique(), "Their count : ",len(dialect_df.dialect.unique())
```

```python
dialect_df.info()
```

```python
dialect_df.groupby('dialect').count()
```

Convert id column into string (since API take it as string)

```python
id_list = list(dialect_df['id'].apply(lambda x: str(x)))
id_list
```

Test first 10 tweets

```python
resp = req.post(API_URL, data=json.dumps(id_list[:10]))
resp.json()
```

check the last 197 id to decide if we can drop them or not

```python
dialect_df[458000:].groupby('dialect').count()
```

They are from only one class

```python
retrived_df = pd.DataFrame(columns=['id','text'])
all_dict = dict()
for i in range(0,len(id_list),1000):
    resp = req.post(API_URL, data=json.dumps(id_list[i:i+1000]))
    if i == 45800:
        resp = req.post(API_URL, data=json.dumps(id_list[i:i+197]))
    all_dict.update(resp.json())
    
print(len(all_dict))
    # break
retrived_df = pd.DataFrame(all_dict.items(),columns=['id','text'])
print(len(retrived_df),len(dialect_df))
retrived_df
```

Good, All text retrived !

```python
#convert type of id col from object into int 
retrived_df.id = retrived_df.id.astype('int64')
```

```python
#check type of the column
retrived_df.id.dtype
```

```python
#merge dialect_df and retrived_df
merged_df = pd.merge(dialect_df,retrived_df,on=['id'])
merged_df
```

## Save data into a file

```python
#make a new csv file from the merged df
merged_df.to_excel('full_dialect_dataset.xlsx',encoding = 'utf8')
```
